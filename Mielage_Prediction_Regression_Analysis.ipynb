**Mileage Prediction**

```python
import pandas as pd
```

```python
mileage = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/MPG.csv')
mileage.head()
```

```python
mileage.info()
```

```python
mileage.describe()
```

```python
mileage = mileage.dropna()
```

```python
mileage.columns
```

```python
y = mileage['mpg']
x = mileage[['displacement', 'horsepower', 'weight', 'acceleration']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

Here we used Linear regression to train this model. This model determines the mileage based on displacement, horsepower, weight and acceleration. Atlast the accuracy is 86%
